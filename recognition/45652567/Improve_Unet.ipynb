```python
inputs = Input(shape=(256,256,1))
conv0 = Conv2D(16, 3, activation = LeakyReLU(alpha=0.01), padding = 'same')(inputs)
conv01 = Conv2D(16, 3, activation = LeakyReLU(alpha=0.01), padding = 'same')(conv0)
drop0 = Dropout(0.3)(conv01)
conv01 = Conv2D(16, 3, activation = LeakyReLU(alpha=0.01), padding = 'same')(drop0)
conv01 = Add()([conv0,conv01])
conv0f = Conv2D(32, 3, activation = LeakyReLU(alpha=0.01), padding = 'same', strides = (2,2))(conv01)

conv1 = Conv2D(32, 3, activation = LeakyReLU(alpha=0.01), padding = 'same')(conv0f)
drop1 = Dropout(0.3)(conv1)
conv1 = Conv2D(32, 3, activation = LeakyReLU(alpha=0.01), padding = 'same')(drop1)
conv1 = Add()([conv0f,conv1])
conv1f = Conv2D(64, 3, activation = LeakyReLU(alpha=0.01), padding = 'same', strides = (2,2))(conv1)

conv2 = Conv2D(64, 3, activation = LeakyReLU(alpha=0.01), padding = 'same')(conv1f)
drop2 = Dropout(0.3)(conv2)
conv2 = Conv2D(64, 3, activation = LeakyReLU(alpha=0.01), padding = 'same')(drop2)
conv2 = Add()([conv1f,conv2])
conv2f = Conv2D(128, 3, activation = LeakyReLU(alpha=0.01), padding = 'same', strides = (2,2))(conv2)

conv3 = Conv2D(128, 3, activation = LeakyReLU(alpha=0.01), padding = 'same')(conv2f)
drop3 = Dropout(0.3)(conv3)
conv3 = Conv2D(128, 3, activation = LeakyReLU(alpha=0.01), padding = 'same')(drop3)
conv3 = Add()([conv2f,conv3])
conv3f = Conv2D(256, 3, activation = LeakyReLU(alpha=0.01), padding = 'same', strides = (2,2))(conv3)

conv4 = Conv2D(256, 3, activation = LeakyReLU(alpha=0.01), padding = 'same')(conv3f)
drop4 = Dropout(0.3)(conv4)
conv4 = Conv2D(256, 3, activation = LeakyReLU(alpha=0.01), padding = 'same')(drop4)
conv4 = Add()([conv3f,conv4])

up1 = UpSampling2D(size = (2,2))(conv4)
conc1 = concatenate([up1, conv3])
conv5 = Conv2D(128, 3, activation = LeakyReLU(alpha=0.01), padding = 'same')(conc1)
conv5 = Conv2D(128, 1, activation = LeakyReLU(alpha=0.01), padding = 'same')(conv5)
up2 = UpSampling2D(size = (2,2))(conv5)
conc2 = concatenate([up2, conv2])

conv6 = Conv2D(64, 3, activation = LeakyReLU(alpha=0.01), padding = 'same')(conc2)
conv6 = Conv2D(64, 1, activation = LeakyReLU(alpha=0.01), padding = 'same')(conv6)
seg1 = Conv2D(4, 3, activation = LeakyReLU(alpha=0.01), padding = 'same')(conv6)
seg1 = UpSampling2D(size= (2,2))(seg1)
up3 = UpSampling2D(size = (2,2))(conv6)
conc3 = concatenate([up3, conv1])

conv7 = Conv2D(32, 3, activation = LeakyReLU(alpha=0.01), padding = 'same')(conc3)
conv7 = Conv2D (32, 1, activation = LeakyReLU(alpha=0.01), padding = 'same')(conv7)
seg2 = Conv2D(4, 3, activation = LeakyReLU(alpha=0.01), padding ='same')(conv7)
up4 = UpSampling2D(size= (2,2))(conv7)
conc4 = concatenate([up4,conv0])

conv8 = Conv2D(32, 3, activation = LeakyReLU(alpha=0.01), padding = 'same')(conc4)

seg3 = Conv2D(4, 3, activation = LeakyReLU(alpha=0.01), padding = 'same')(conv8)

sum1 = Add()([seg1, seg2])
sum1 = UpSampling2D(size= (2,2))(sum1)
sum2 = Add()([sum1, seg3])

output = Conv2D(4, 1, activation = 'sigmoid')(sum2)

model = Model(inputs=inputs, outputs=output)
```

```python
model.compile(optimizer = opti, loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
def dice_coef(y_true, y_pred):
    
    intersection = tf.reduce_sum(y_true * y_pred, axis=[1,2])
    joint = tf.reduce_sum(y_true, axis=[1,2]) + tf.reduce_sum(y_pred, axis=[1,2])
    loss = (2. * intersection) / (joint + 1e-6)
    return tf.reduce_mean(loss, axis=0)
    
```
