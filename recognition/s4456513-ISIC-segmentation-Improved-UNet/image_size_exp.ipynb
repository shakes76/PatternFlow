```python
import glob as gb
import cv2
```

```python
image_size_dic  = {}


image_path =  "ISIC/ISIC2018_Task1-2_Training_Input_x2/*.jpg"
# mask_path = "ISIC/ISIC2018_Task1_Training_GroundTruth_x2/*.png"

images = sorted(gb.glob(image_path))
# masks = sorted(gb.glob(mask_path))

index = 0
for image_path in images:
#     img = tf.io.read_file(image_path)
#     img = tf.image.decode_jpeg(img, channels = 1)
#     size = img.shape.as_list
#     print(size)

    img = cv2.imread(image_path)
    h, w, d = img.shape
    size = (w,h)
    if size in image_size_dic:
        image_size_dic[size] += 1
    else:
        image_size_dic[size] = 1
for  key in image_size_dic:
    print('Size: ', key)
    print('Number of images with this size: ', image_size_dic[key])
    print("#------------------------------------------------#")
```

```python
sum = 0
size_dic = {256: 0, 512:0, 1024:0}
for key in image_size_dic:
    sum += image_size_dic[key]
    w, h = key
    if w > 1024:
        size_dic[1024] += image_size_dic[key]
    elif w > 512:
        size_dic[512] += image_size_dic[key]
    else:
        size_dic[256] += image_size_dic[key]

print("Width Distribution of Raw Images")
print("Range 0 - 512:    ", size_dic[256])
print("Range 512 - 1024: ", size_dic[512])
print("Range > 1024:     ", size_dic[1024])
    
    
```
